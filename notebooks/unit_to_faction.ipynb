```python
import pandas as pd

# --- load main_units
main_units = pd.read_csv(
    r"..\data\raw\main_units_tables.tsv",
    sep="\t", dtype=str, keep_default_na=False
)
main_units = main_units[~main_units["unit"].str.startswith("#")]  # <-- IMPORTANT

all_units = set(main_units["unit"].unique())

# --- load u2f
u2f_all = pd.read_csv(
    r"..\data\raw\units_to_exclusive_faction_permissions_tables.tsv",
    sep="\t", dtype=str, keep_default_na=False
)
u2f_all = u2f_all[~u2f_all["key"].str.startswith("#")]

u2f_allowed = u2f_all[u2f_all["allowed"].str.lower() == "true"]
factions = sorted(u2f_allowed["faction"].unique())

restricted_units = set(u2f_all["key"].unique())
unrestricted_units = all_units - restricted_units

def full_roster_for_faction(faction_key: str) -> list[str]:
    allowed_for_faction = set(
        u2f_allowed.loc[u2f_allowed["faction"] == faction_key, "key"].unique()
    )
    return sorted(unrestricted_units | allowed_for_faction)

# --- test correctly
target_faction = "wh2_dlc09_skv_clan_rictus"
roster = full_roster_for_faction(target_faction)

print(target_faction, "roster size:", len(roster))
print(roster[:50])
for u in roster:
    print(u)

```
