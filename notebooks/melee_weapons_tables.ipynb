```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
melee_weapons =     pd.read_csv(r"..\data\raw\melee_weapons_tables.tsv",        sep="\t", dtype=str, keep_default_na=False)
```

```python
melee_weapons.info()
```

```python
melee_weapons["melee_attack_interval"].unique()
```

```python
melee_weapons["melee_attack_interval"].value_counts()
```

```python
vc = melee_weapons["melee_attack_interval"].value_counts().sort_index()
ax = vc.plot(kind="bar", figsize=(12,4))

# only label bars < 100
labels = [f"{int(v)}" if v < 100 else "" for v in vc.values]
ax.bar_label(ax.containers[0], labels=labels, padding=3, fontsize=9)

ax.set_xlabel("melee_attack_interval")
ax.set_ylabel("count")
ax.set_title("Distribution of melee weapon attack speed interval")

plt.show()
```
